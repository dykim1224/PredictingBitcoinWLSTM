```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from wordcloud import STOPWORDS
```

```python
file_path = r"C:\Users\jhkju\Documents\GoogleTrend_RelatedQuries2.csv"

df = pd.read_csv(file_path)
df
```

```python
rising = df['RISING']
text_data = " ".join(rising.astype(str))

stop_words = set(STOPWORDS)
custom_stopwords = {"https", "co", "RT", "will", "el", "news", "sec", "date", "time", "un"} 
stop_words.update(custom_stopwords)

wordcloud = WordCloud(
    width=800,
    height=400,
    max_words=50,
    background_color="white",
    stopwords=stop_words,
    collocations=False,
    normalize_plurals=False,
    contour_color="steelblue",
    contour_width=1,
    scale=2
).generate(text_data)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Related Quries With Bitcoin", fontsize=16)
plt.show()
```
